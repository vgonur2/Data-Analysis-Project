The code above basically goes through how we removed the na values, we did .dropna() to drop all na values, and redid df.isna().sum() to confirm the values actually got dropped, this left us with 948 rows (dropped 264 rows) as we can see in the third code box. 

One caveat with the method we used is that it dropped all values the computer deemed as na, however, there might be implicit na values that we wont know are na values such as "MISSING" or "Not Found", these values are not found and recongnized by the computer, and becuase the dataset is so large that we can not comb through it by hand, we can not find implicit na values. 



```python
pub_counts = df["Metadata.Publishers"].value_counts()
```

```python
pub_counts >= 10
```

```python
pub_counts_new = pub_counts[pub_counts >= 10]
pub_counts_new
```

```python
top_pub_counts = pub_counts_new.index
top_pub_counts
```

```python
df = df[df['Metadata.Publishers'].isin(top_pub_counts)]
df.shape[0]
```

In the code above, we removed all publishers in the dataset, that did not have a value count of more than 10, this was done, to only include notable publishers to better help our analysis, and not have multiple publishers with low value counts, who could maybe cause our analysis to weaken.

We dropped 20 more rows
